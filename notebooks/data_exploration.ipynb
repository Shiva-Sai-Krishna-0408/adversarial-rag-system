```python
from pypdf import PdfReader
import os
import re
```

```python


def extract_text_pypdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""

    for pages in reader.pages:
        extracted = pages.extract_text()
        text += extracted
    return text 

path = r'C:\Users\Shiva\projects\adversarial-rag-system\data'
content = os.listdir(path)
print(content)

```

```python
def chunking(text, chunk_size, overlap):
    words = text.split()
    start = 0
    end = start + chunk_size
    chunks = []

    while start < len(words):
        words_500 = words[start:end]
        start += chunk_size - overlap
        end  = start + chunk_size
        words_str = ' '.join(words_500)
        chunks.append(words_str)

    return chunks      
```

```python

for c in content:
    file_path = os.path.join(path, c)
    files = ['apple_tos.pdf','github_tos.pdf','openai_tos.pdf']
    if c in files:
        t =  extract_text_pypdf(file_path)
        words_1 = t.split()
        words = ' '.join(words_1)
        #print(words[:5000])
        counter = 0
        clean = []
        temp = []
        for w in words.split():
            if len(w) > 1:
                if temp:
                    clean.extend(temp)
                    temp.clear()
                clean.append(w)
                counter = 0
            else:
                counter += 1
                temp.append(w)
                if counter > 5:
                    temp.clear()
        clean_data = ' '.join(clean)
        #print(clean_data[:5000])
        clean_1 = re.sub('\d+/\d+/\d+,\s\d+:\d+\s[AP]M','',clean_data)
        clean_2 = re.sub('https://\S+','',clean_1)
        clean_2 = clean_2.replace('Legal - iCloud - Apple','')
        clean_3 = re.sub('\d+/\d+','',clean_2)
        chunking(clean_3, 500, 50)

```
